```python
import numpy as np
import cv2
from glob import glob
import matplotlib.pyplot as plt
```

```python
fpath = glob('./data/female/*.jpg')
mpath = glob('./data/male/*jpg')
```

```python
print('Total number of images in female folder', len(fpath))
print('Total number of images in male folder', len(mpath))
```

```python
### Reading Image

img = cv2.imread(fpath[0]) #image is in GBR format in opencv
```

```python
### Converting image to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()
```

```python
### Apply haar cascade classifier
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
```

```python
### Converting image to grayscale
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
plt.show()
```

```python
face_list = haar.detectMultiScale(gray, 1.5, 5)
print(face_list)
```

```python
###Draw bounding box using the coordinates

for x,y,w,h in face_list:
    cv2.rectangle(img_rgb, (x,y), (x+w, y+h), (0,255,0), 3)
    
    #crop the detected image
    crp_img = img_rgb[y:y+h, x:x+h]   #give row and column coordinates
    plt.imshow(crp_img)
    plt.show()
    
    
```
